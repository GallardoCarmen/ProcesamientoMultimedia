```python
cascade = cv2.CascadeClassifler("haarcascade_frontalface_alt.xml")

cap = cv2.open("face.mp4")

while True:
  frame, _ = cap.read()

  gris = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

  faces = cascade.detectMultiScale(gris, 1.2,4)

  for (x,y,w,h) in faces:
    cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 3);

  jcv2.imshow("faces", frame)


  if cv2.waitKey(25) & 0xFF == ord('q'):
    break

```
