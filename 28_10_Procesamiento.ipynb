```python
import cv2

from google.colab import output
output.enable_custom_widget_manager()

from google.colab.patches import cv2_imshow

from matplotlib import pyplot as plt
import opencv_jupyter_ui as jcv2
import numpy as np
```

```python
Mario = cv2.imread("mario.png")
jcv2.imshow("", Mario)

mariogray = cv2.cvtColor(Mario, cv2.COLOR_BGR2GRAY);

cv2_imshow(mariogray)

_,maribobin = cv2.threshold(mariogray, 100, 200, cv2.THRESH_BINARY)

cv2_imshow(maribobin)

marioblur = cv2.blur(Mario, [16,16])

cv2_imshow(marioblur)

mariogray_display = cv2.cvtColor(mariogray,cv2.COLOR_GRAY2BGR)

cv2_imshow(mariogray_display)
#morfologiex dilate y tal para opens ycclose
```

```python
coins = cv2.imread("coins.jpg")
jcv2.imshow("", coins)
coinsgray = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY);

coinsgray_display = cv2.cvtColor(coinsgray,cv2.COLOR_GRAY2BGR)

cv2_imshow(coinsgray_display)


_,coinsbinary = cv2.threshold(coinsgray, 150, 255, cv2.THRESH_BINARY)
coinsbinary_display = cv2.cvtColor(coinsbinary,cv2.COLOR_GRAY2BGR)

cv2_imshow(coinsbinary_display)
```

```python
brush = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))

coinsOpen = cv2.morphologyEx(coinsbinary, cv2.MORPH_OPEN,  brush)
brush = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
coinsClose = cv2.morphologyEx(coinsOpen, cv2.MORPH_CLOSE,  brush)

coinsClose_display = cv2.cvtColor(coinsClose,cv2.COLOR_GRAY2BGR)

jcv2.imshow("", coinsClose_display)
```

```python
clusters, _ = cv2.findContours(coinsClose, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

countCoins = 0
for cluster in clusters:
  area  = cv2.contourArea(cluster)

  areamin = 300


  if area > areamin:
    cv2.drawContours(coinsClose_display, [cluster], -1, (0,255,0), 5)
    jcv2.imshow("", coinsClose_display)
    countCoins += 1

jcv2.imshow("",coinsClose_display)
print(countCoins)
```
