```python
import cv2

from google.colab import output
output.enable_custom_widget_manager()

from google.colab.patches import cv2_imshow

from matplotlib import pyplot as plt
import opencv_jupyter_ui as jcv2
import numpy as np
```

```python
Mario = cv2.imread("mario.png")
jcv2.imshow("", Mario)

mariogray = cv2.cvtColor(Mario, cv2.COLOR_BGR2GRAY);

cv2_imshow(mariogray)

_,maribobin = cv2.threshold(mariogray, 100, 200, cv2.THRESH_BINARY)

cv2_imshow(maribobin)

marioblur = cv2.blur(Mario, [16,16])

cv2_imshow(marioblur)
#morfologiex dilate y tal para opens ycclose
```
